```python
def get_recom(stock_code, sdate, edate):
    df1 = yf.download(
                        tickers = stock_code,
                        start=sdate,
                        #end=date.today().replace(day=2),
                        end = edate,
                        interval = "1d",
                        threads=True,
                        group_by = 'ticker',
                        progress=False
                    )


    df1 = df1[['Adj Close', 'Volume']]
    df1.columns = ['price', 'vol']

    

    df1 = df1.reset_index()

    avg = df1['price'].mean()
    #print(avg)

    df1['Week_Number'] = df1['Date'].dt.isocalendar().week

    list_t = []


    for i in list(set(df1['Week_Number'])):
        try:
            index = df1.loc[df1['Week_Number'] == i, 'Date'].tail(1).index[0]
            last_day_week = df1.loc[df1['Week_Number'] == i, 'Date'].tail(1)[index].strftime('%Y-%m-%d')

            corr_val = df1.loc[df1['Week_Number'] == i,:].corr(numeric_only = True).iloc[0:1, 1].values[0]


            price_list = list(df1.loc[df1['Week_Number'] == i, 'price'])
            pc = round(((price_list[-1] - price_list[0])/price_list[0])*100)

            list_t.append((last_day_week, corr_val,  pc))


        except:
            continue

    df = pd.DataFrame(list_t, columns =['date', 'corr', 'pc'])
    #print(df)

    # for buying consecutive -ve correlation with dropping prices 
    # surely negative but good price to buy
    # imp date - 2 consecutive -ve days
    # v_imp_date - 3 consecutive -ve days


    imp_date = []
    v_imp_date = []

    
    try:
        for i in range(len(df)):
            if ((df.loc[i, 'corr'] < 0) &(df.loc[i, 'pc'] <= 0)):
                if ((df.loc[i+1, 'corr'] < 0) &(df.loc[i+1, 'pc'] <= 0)):
                    imp_date.append(df.loc[i+1, 'date'])
                    if ((df.loc[i+2, 'corr'] < 0) &(df.loc[i+2, 'pc'] <= 0)):
                        v_imp_date.append(df.loc[i+2, 'date'])

        dates = list(set(imp_date) | set(v_imp_date))
    except:
        dates = []
    #print(dates)

    df = df[df['corr'] < -0.05]

    df['index'] = df['corr']*df['pc']
    #print(df.sort_values(by = 'index', ascending= False).head(2))

    dates = sorted(dates, key=lambda x: datetime.strptime(x, '%Y-%m-%d'))

    return df1, dates

    #return df
```
